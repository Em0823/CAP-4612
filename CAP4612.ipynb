```python
import pandas as pd
import numpy as np 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error

from pandas import DataFrame
```

```python
car = pd.read_csv('https://raw.githubusercontent.com/Em0823/CAP-4612/main/car_prediction_data1.csv')
car
```

```python
car_features = car[['engine', 'hp', 'city_mpg']]

car_labels = car[['price']]

print(car_features)
print(car_labels)
```

```python

car_features_train, car_features_test, car_label_train, car_label_test = train_test_split(car_features, car_labels, random_state=0, train_size = .7)
```

```python
knn = KNeighborsClassifier( n_neighbors= 7, p = 2)

print(knn.get_params())
knn.fit(car_features_train, np.ravel(car_label_train))

prediction = knn.predict(car_features_test)

prediction
```

```python
accuracy = accuracy_score(car_label_test, prediction)
from sklearn.metrics import recall_score
recall = recall_score(car_label_test, prediction, average='macro', zero_division = 1)
metrics = precision_recall_fscore_support(car_label_test, prediction, average='macro', zero_division = 1)
metrics
```

```python
MSE  = mean_squared_error(car_label_test, prediction)
RMSE = mean_squared_error(car_label_test, prediction, squared=False)
print("MSE: %5.3f.   RMSE: %5.3f" %(MSE, RMSE))
```
